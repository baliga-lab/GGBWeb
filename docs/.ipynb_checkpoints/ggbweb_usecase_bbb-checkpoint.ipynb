**Nucleotide metabolism**
- GO:0009117 - nucleotide metabolic process

```python
nucleotide_genes = pd.DataFrame( list( client[ db ].row_info.find( { "$or": [ { "GO": { "$regex" : "GO:0009117" } }, { "TIGRRoles": { "$regex" : "nucleotide" } }, { "ECDesc": { "$regex" : "nucleotide" } } ] } ) ) )
nucleotide_genes.loc[ :,[ "name","ECDesc","TIGRRoles" ] ]
```

```python
# name chromosome-:1823979-1824947 
ggbwebModule( genes = nucleotide_genes.name, outfile = "nucleotide_module.txt", host = host, port = port, db = db, input_type = "name" )
```
